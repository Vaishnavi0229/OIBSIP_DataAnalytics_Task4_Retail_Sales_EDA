# Retail Sales EDA #

# Data Loading and Cleaning #

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\vemul\OneDrive\Desktop\Internship Projects\retail_sales_dataset.csv")

# Check data
df.head()
df.info()

# Cleaning
df.drop_duplicates(inplace=True)
df.ffill(inplace=True)

# Convert date column
df['Date'] = pd.to_datetime(df['Date'])
```

# Descriptive Statistics #

```python
import matplotlib.pyplot as plt

sales_trend = df.groupby('Date')['Total Amount'].sum()

plt.figure(figsize=(12,5))
sales_trend.plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

# Time Series Analysis #

```python
mean_sales = df['Total Amount'].mean()
median_sales = df['Total Amount'].median()
mode_sales = df['Total Amount'].mode()[0]
std_sales = df['Total Amount'].std()

print("Mean:", mean_sales)
print("Median:", median_sales)
print("Mode:", mode_sales)
print("Standard Deviation:", std_sales)
```

# Customer Product Analysis #

```python
# Product category analysis
category_sales = df.groupby('Product Category')['Total Amount'].sum()

# Customer demographics
gender_sales = df.groupby('Gender')['Total Amount'].sum()

print(category_sales)
print(gender_sales)
```

# Visualizations #

```python
import seaborn as sns

# Bar chart
category_sales.plot(kind='bar', title='Sales by Product Category')

# Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
